```python
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
import random
NumOfPoints1=12
NumOfPoints2=10
x1=[]
y1=[]
xmin1,xmax1=0,6
ymin1,ymax1=0,6
xmin2,xmax2=5,10
ymin2,ymax2=5,10
for i in range(NumOfPoints1):
    x1.append(random.uniform(xmin1,xmax1))
    y1.append(random.uniform(ymin1,ymax1))
for i in range(NumOfPoints2):
    x1.append(random.uniform(xmin2,xmax2))
    y1.append(random.uniform(ymin2,ymax2))
print(x1)
print(y1)
k=2
xc1=random.uniform(0,5)
yc1=random.uniform(0,5)
xc2=random.uniform(5,10)
yc2=random.uniform(5,10)
def klaster(xc1,yc1,xc2,yc2):
    dist1=[]
    dist2=[]
    for i in range(NumOfPoints1 + NumOfPoints2):
        dist1.append(sqrt((x1[i]-xc1)**2+(y1[i]-yc1)**2))
        dist2.append(sqrt((x1[i]-xc2)**2+(y1[i]-yc2)**2))
    metka=[]
    for j in range(NumOfPoints1 + NumOfPoints2):
        if dist1[j]> dist2[j]:
            metka.append(2)
        else:
            metka.append(1)
    
    nclass1=metka.count(1)
    nclass2=metka.count(2)
    srx1=0
    sry1=0
    srx2=0
    sry2=0
    for k in range(nclass1+nclass2):
        if metka[k]==1:
            srx1+=x1[k]
            sry1+=y1[k]
        else:
            srx2+=x1[k]
            sry2+=y1[k]
    srx1=srx1/nclass1
    sry1=sry1/nclass1
    srx2=srx2/nclass2
    sry2=sry2/nclass2
    return float(srx1),float(sry1),float(srx2),float(sry2),metka

a,b,c,d,m=xc1,yc1,xc2,yc2,[0]
for i in range(2):
    a1,b1,c1,d1,m=klaster(a,b,c,d)
    a,b,c,d,m=klaster(a1,b1,c1,d1)
print(a,b,c,d,m)
class1x=[]
class1y=[]
class2x=[]
class2y=[]
for i in range(len(m)):
    if m[i]==1:
        class1x.append(float(x1[i]))
        class1y.append(float(y1[i]))
    else:
        class2x.append(float(x1[i]))
        class2y.append(float(y1[i]))
r1=0
dist1=[]
dist2=[]
for i in range(len(class1x)):
    dist1.append(sqrt((class1x[i]-a)**2+(class1y[i]-b)**2))
for i in range(len(class2x)):
    dist2.append(sqrt((class2x[i]-c)**2+(class2y[i]-d)**2))
fig,ax=plt.subplots()
circle1 = plt.Circle((a, b), max(dist1), color='r', alpha=0.3, fill=False)
circle2 = plt.Circle((c, d), max(dist2), color='b', alpha=0.3, fill=False)


ax.add_patch(circle1)
ax.add_patch(circle2)
plt.scatter(x1,y1)

plt.show()
```
